# Evaluate the dataset of the Summer Olympics, London 2012

```python
# importing pandas as pd
import pandas as pd
```

```python
df = pd.read_excel("/home/user/Downloads/Olympic 2012 Medal Tally.xlsx") # reading dataset
df.head()
```

## Cleaning the data

```python
df=df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df.columns
```

```python
# renaming columns
df=df.rename(columns={'Unnamed: 1':'Country', 'Unnamed: 2':'Country Code', 'Unnamed: 3':'Year', 
                        'Unnamed: 4':'Medal Tally: Gold', 'Unnamed: 5':'Medal Tally: Silver', 'Unnamed: 6':'Medal Tally: Bronze'})
df
```

```python
# removing first two rows
df=df.drop([0,1], axis=0)
df
```
